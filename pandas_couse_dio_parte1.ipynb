# **Python para análise de dados(Pandas)**

```python
# Fazendo a import do Pandas e colocando um apelido de pd
import pandas as pd
```

```python
# Lendo o dataframe
df = pd.read_csv('/content/drive/MyDrive/datasets/Gapminder.csv', error_bad_lines=False, sep=";") # Ignorando as linhas com erro e definindo que o sep é por ;
```

```python
# Visualizando as 5 primeiras linhas
df.head()
```

```python
# Renomeando as colunas

df = df.rename(columns={"country": "pais", "continent": "continente", "year": "ano", "lifeExp": "expectativaVida", "pop": "popTotal", "gdpPercap": "PIB"})
```

```python
df.head(10) # Mostrando as 10 primeiras linhas
```

```python
df.shape # Mostrando a quantidade de linhas e colunas
```

```python
df.columns # Retornando o nome das colunas na base de dados
```

```python
df.dtypes # Mostrando o tipo de dados armazenados em cada coluna
```

```python
df.tail(15) # Mostrando as 15 últimas linhas do dataframe
```

```python
df.describe() # Estatísticas gerais do dataframe
```

```python
# Visualizando os continentes disponíveis
df["continente"].unique() # Mostrando os valores únicos da coluna continente
```

```python
# Retornando os valores somente do continente "oceania"

oceania = df.loc[df["continente"] == "Oceania"] # Colocando em uma variável as linhas com o continete == Oceania
oceania.head() # Mostrando os 5 primeiros
```

```python
# Confirmando que só foi retornado com o continente "Oceania"
oceania["continente"].unique()
```

```python
# Fazendo a contagem de países por cada continente usando o groupby
df.groupby("continente")["pais"].nunique()
```

```python
# Vendo a expectativa de vida média por cada ano de nascimento
df.groupby("ano")["expectativaVida"].mean()
```

```python
# Média do PIB, já aparece no describe
df['PIB'].mean()
```

```python
# Retornando a soma dos valores do PIB
df['PIB'].sum()
```
