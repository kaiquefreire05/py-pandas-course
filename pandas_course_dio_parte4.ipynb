```python
# Criando nosso DataFrame
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
# Visualizando 5 primeiras linhas
df.head()
```

```python
# Quantidade de linhas e colunas
df.shape
```

```python
# Verificando os tipos de dados
df.dtypes
```

```python
# Receita Total
df['Valor Venda'].sum()
```

```python
# Criando a coluna Custo Total
df['Custo'] = df['Custo Unit치rio'].mul(df['Quantidade']) # Criando a coluna de custo
df.head(1)
```

```python
# Soma do Custo Total
round(df['Custo'].sum(), 2) # Usando o round para printar apenas o valor em 2 casas decimais
```

```python
# Tenho o valor de venda e o custo do produto, ent칚o vou criar uma coluna para visualizar o lucro dos produtos
df['Lucro'] = df['Valor Venda'] - df['Custo']
df.head(2)
```

```python
# Total de lucro
round(df['Lucro'].sum(), 2)
```

```python
df['Tempo Envio'] = df['Data Envio'] - df['Data Venda']
df.head(2)
```

**Agora, quero saber o a m칠dio do tempo de envio para cada marca, primeiro passo 칠 transformar a coluna 'Tempo Envio' em num칠rica**

```python
df['Tempo Envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
df.head(1)
```

```python
# Pegando a m칠dia de tempo de envio por 'Marca'
df.groupby('Marca')['Tempo Envio'].mean() # Fazendo um agrupamento por marca para saber a m칠dio do tempo de envio
```

```python
# Verificando se existe valores ausentes
df.isnull().sum() # Fazendo a soma dos valores ausentes no dataframe
```

```python
# Fazendo a convers칚o dos n칰meros da nota칞칚o cient칤fica
pd.options.display.float_format = '{:20,.2f}'.format

# Fazendo agrupamento por Ano e Marca
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
# Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, 'Marca'])['Lucro'].sum().reset_index()
lucro_ano
```

```python
# Total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Gr치fico total de vendas por produto
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total de produtos vendidos')
plt.xlabel('Total')
plt.ylabel('Produto')
```

```python
# Fazendo agrupamento por Ano e Marca
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro X Ano')
plt.xlabel('Ano')
plt.ylabel('Lucro')
```

```python
# Selecionando apenas as vendas em 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
df_2009.head(2)
```

```python
# Pegando apenas o lucro por m칡s no ano de 2009
df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title='Lucro X M칡s')
plt.xlabel('M칡s')
plt.ylabel('Lucro')
```

```python
# Pegando apenas o lucro por marca no ano de 2009
df_2009.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro X Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
# Pegando apenas o lucro por classe no ano de 2009
df_2009.groupby('Classe')['Lucro'].sum().plot.bar(title='Lucro X Classe')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
